# Random inspections of other datasets

### Boat files inspection
 Get min and max file lengths + sample rate. For summary csv.

```python
boats_folder = r'C:\Users\ben07\OneDrive - University College London\UCL PhD\Random stuff\Clemency boats\all_files'

import wave
import os

def get_wav_info(wav_file_path):
    # Open the .wav file
    with wave.open(wav_file_path, 'rb') as wf:
        # Get the sample rate
        sample_rate = wf.getframerate()
        # Get the total number of frames
        n_frames = wf.getnframes()
        # Calculate the duration of the audio in seconds
        duration = n_frames / sample_rate

    return sample_rate, duration

# Initialize variables for min and max file length
min_length = float('inf')
max_length = float('-inf')

# Loop through all files in the folder
for filename in os.listdir(boats_folder):
    if filename.endswith('.wav'):
        filepath = os.path.join(boats_folder, filename)
        sample_rate, duration = get_wav_info(filepath)

        print(f"File: {filename}, Sample Rate: {sample_rate}, Duration: {duration} seconds")

        # Update min and max lengths
        min_length = min(min_length, duration)
        max_length = max(max_length, duration)

print(f"Min file length: {min_length} seconds")
print(f"Max file length: {max_length} seconds")
```
